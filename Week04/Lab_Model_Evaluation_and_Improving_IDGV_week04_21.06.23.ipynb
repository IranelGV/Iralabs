```python
from sklearn.preprocessing import OrdinalEncoder

education_hierarchy = ['high school or below', 'college', 'bachelor', 'master', 'doctor']
coverage_hierarchy = ['basic', 'extended', 'premium']
policy_hierarchy = ['personal l1', 'personal l2', 'personal l3', 'corporate l1', 'corporate l2','corporate l3', 'special l1', 'special l2', 'special l3']
renew_offer_type_hierarchy = ['offer1', 'offer2', 'offer3', 'offer4']
vehicle_size_hierarchy = ['small', 'medsize', 'large']
```

```python
encoder1 = OrdinalEncoder(categories=[education_hierarchy])
encoder2 = OrdinalEncoder(categories=[coverage_hierarchy])
encoder3 = OrdinalEncoder(categories=[policy_hierarchy])
encoder4 = OrdinalEncoder(categories=[renew_offer_type_hierarchy])
encoder5 = OrdinalEncoder(categories=[vehicle_size_hierarchy])
```

```python
ordinal_encoded = pd.DataFrame({
    'education_encoded': encoder1.fit_transform(df[['education']]).flatten().astype(int),
    'coverage_encoded': encoder2.fit_transform(df[['coverage']]).flatten().astype(int),
    'policy_encoded': encoder3.fit_transform(df[['policy']]).flatten().astype(int),
    'renew_offer_type_encoded': encoder4.fit_transform(df[['renew_offer_type']]).flatten().astype(int),
    'vehicle_size_encoded': encoder5.fit_transform(df[['vehicle_size']]).flatten().astype(int)
})
```

```python
ordinal_encoded
```

```python
categorical_converted = pd.concat([ordinal_encoded,nominals_converted], axis=1)
```

```python
numerical_categorical_converted = pd.concat([categorical_converted, numerical_df], axis =1)
```
