```python
def lower_case_column_names(df):
    df.columns=[i.lower() for i in df.columns]
    return df
```

```python
def unique_values(df):
    df.columns = list(map(lambda x: x.replace(" ","_") if type(x)==str else x , df.columns ))
    return df.columns
```

```python
unique_values(df)
```

```python
def get_unique_values(df):
    unique_values = {}
    
    for column in df.columns:
        unique_values[column] = df[column].unique()
    
    return unique_values
```

```python
def fill_nulls_values(df):
    
    #float_columns = df.select_dtypes(include=['float64'])
    
    float_columns = df.select_dtypes(include=[float])
    
    mean_float = float_columns.mean()
    
    
    fill_nulls_mean= df.fillna(mean_float, inplace=True)
    
    return fill_nulls_mean
```

```python
fill_nulls_values(df)
```

```python
def fill_nulls_values_mode(df):
    
    object_columns = df.select_dtypes(include=['object'])
    
    mode_object = object_columns.mode().values[0]
    
    
    fill_nulls_mode_ob= df.fillna(mode_object, inplace=True)
    
    return fill_nulls_mode_ob
```

```python
fill_nulls_mode_ob
```

```python
def fill_nulls_values_str(df):
    
    str_columns = df.select_dtypes(include=['int64'])
    
    mode_str = str_columns.mode().values[0]
    
    
    fill_nulls_mode_int= df.fillna(mode_str, inplace=True)
    
    return fill_nulls_mode_int
```
